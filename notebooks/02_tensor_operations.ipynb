# Tensor operations

Reshape, slice, broadcast, and combine tensors while checking shape at every step.

```python
import torch

x = torch.arange(24).reshape(2, 3, 4)
x
```

## `view` and `reshape`

`view` requires contiguous storage. `reshape` may copy when it cannot return a view.

```python
matrix = x.reshape(6, 4)
transposed = matrix.T
print(matrix.is_contiguous(), transposed.is_contiguous())
print(transposed.reshape(2, 12).shape)
```

## Slicing and broadcasting

```python
first_channel = x[:, :, 0]
offsets = torch.tensor([0, 10, 20, 30])
shifted = x + offsets
first_channel, shifted[0]
```

## Concatenate versus stack

```python
a = torch.ones((2, 3))
b = torch.zeros((2, 3))
print("cat", torch.cat((a, b), dim=0).shape)
print("stack", torch.stack((a, b), dim=0).shape)
```
