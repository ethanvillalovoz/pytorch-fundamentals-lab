# Tensors

Start with shape, dtype, and device. Those three properties explain most tensor errors before a model is involved. Install the lab first with `pip install -e .`.

```python
import torch

torch.manual_seed(42)
```

```python
features = torch.tensor([[5.1, 3.5, 1.4, 0.2], [6.7, 3.0, 5.2, 2.3]])
print(features)
print(f"shape={tuple(features.shape)}, dtype={features.dtype}, device={features.device}")
```

## Construct deliberately

Use constructors that make intent visible. Model inputs are usually floating point; class labels passed to cross entropy are integer indices.

```python
inputs = torch.zeros((3, 4), dtype=torch.float32)
labels = torch.tensor([0, 1, 2], dtype=torch.long)
random_batch = torch.rand((3, 4))
inputs.shape, labels.dtype, random_batch.mean()
```

## Move data and models together

The package uses one device resolver so examples do not silently mix CPU and GPU tensors.

```python
from pytorch_lab.reproducibility import resolve_device

device = resolve_device()
features_on_device = features.to(device)
print(device, features_on_device.device)
```
