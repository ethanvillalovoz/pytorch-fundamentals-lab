# Tensor math and autograd

PyTorch records operations on tensors with `requires_grad=True`, then applies the chain rule when `backward()` is called.

```python
import torch

weight = torch.tensor(2.0, requires_grad=True)
feature = torch.tensor(3.0)
target = torch.tensor(10.0)
prediction = weight * feature
loss = (prediction - target) ** 2
loss.backward()
print(f"loss={loss.item():.1f}, dloss/dweight={weight.grad.item():.1f}")
```

## A complete update

Optimizers zero gradients, backpropagate one scalar loss, then update parameters.

```python
model = torch.nn.Linear(1, 1)
optimizer = torch.optim.SGD(model.parameters(), lr=0.05)
criterion = torch.nn.MSELoss()
x = torch.tensor([[1.0], [2.0], [3.0]])
y = 2 * x + 1

for _ in range(100):
    optimizer.zero_grad(set_to_none=True)
    batch_loss = criterion(model(x), y)
    batch_loss.backward()
    optimizer.step()

with torch.inference_mode():
    print(model(torch.tensor([[4.0]])).item())
```

## Inference mode

Use `torch.inference_mode()` for evaluation. It avoids building a graph and makes the intent explicit.
