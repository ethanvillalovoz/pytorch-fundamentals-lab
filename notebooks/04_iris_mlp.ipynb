# Iris multilayer perceptron

This notebook calls the same deterministic experiment that the CLI and test suite use. The scaler is fit on training data only and the split is stratified.

```python
from pytorch_lab.iris import IrisConfig, train_iris

run = train_iris(IrisConfig(seed=42, epochs=250))
print(f"test accuracy: {run.test_accuracy:.1%}")
print(f"test loss: {run.test_loss:.4f}")
print(run.confusion)
```

## Inspect the learning curve

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7, 3.5))
ax.plot(run.train_loss, color="#2563eb")
ax.set(xlabel="epoch", ylabel="cross entropy", title="Iris MLP training loss")
ax.spines[["top", "right"]].set_visible(False)
plt.show()
```

## Inspect individual predictions

```python
for label, prediction in zip(run.split.test_labels.tolist(), run.predictions.tolist(), strict=True):
    print(
        f"label={run.split.class_names[label]:10s} prediction={run.split.class_names[prediction]}"
    )
```
